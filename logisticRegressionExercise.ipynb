```python
import pandas as pd

data = pd.read_csv("HR_comma_sep.csv")
data
```

```python
data["Department"].unique()
```

```python
# converting the departement and salary columns into integer form as they are in object

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data["el_department"] = le.fit_transform(data["Department"])
data["el_salary"] = le.fit_transform(data["salary"])
data
```

```python
new_data = data.drop(["Department", "salary"], axis = "columns")
new_data["el_salary"].unique()
```

```python
features = new_data.drop("el_salary", axis = "columns")
target = new_data["el_salary"]
target
```

```python
# doing train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
len(X_test)
```

```python
# applying logistic regression

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# checking accuracy of the model

model.score(X_test, y_test)
```
