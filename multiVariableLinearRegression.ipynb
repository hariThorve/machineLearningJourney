```python
import pandas as pd

data = pd.read_csv("hiring.csv")
data
```

```python
# checking whether the data contains any empty values

data.isnull().sum()
```

```python
# filling the null value in experience column through backword filling

data["experience"] = data["experience"].bfill()
data
```

```python
# filling the null value in test_score(out of 10) column through calculating mean

mean_val = data["test_score(out of 10)"].mean()
round_mean_val = round(mean_val,2)
data["test_score(out of 10)"] = data["test_score(out of 10)"].fillna(round_mean_val)
data
```

```python
# next step to convert the string values to integer as our model understands only numbers

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data["en_experience"] = le.fit_transform(data["experience"])
data
```

```python
# In the above data the target is salary and other columns are features

features = data.drop(columns = ["experience", "salary($)"],inplace = False)
target = data["salary($)"]
```

```python
# doing the train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
X_train.shape
```
