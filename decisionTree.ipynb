```python
inputs = data.drop("salary_more_then_100k", axis ="columns")
target = data["salary_more_then_100k"]
```

```python
from sklearn.preprocessing import LabelEncoder

company_en = LabelEncoder()
job_en = LabelEncoder()
degree_en = LabelEncoder()
```

```python
inputs["company_en"] = company_en.fit_transform(inputs["company"])
inputs["job_en"] = job_en.fit_transform(inputs["job"])
inputs["degree_en"] = degree_en.fit_transform(inputs["degree"])
```

```python
inputs_en = inputs.drop(["company", "job", "degree"], axis="columns")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_en, target, test_size=0.2)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()

model.fit(inputs_en, target)
```

```python
model.score(inputs_en, target)
```
